```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
heart_dataset = pd.read_csv('/content/heart.csv')
```

```python
heart_dataset.head()
```

```python
heart_dataset.shape
heart_dataset.describe()
```

```python
heart_dataset.isnull().sum()
```

```python
X =heart_dataset.drop(columns='target', axis =1)
```

```python
Y=heart_dataset['target']
```

```python
X_train, X_test,Y_train,Y_test =train_test_split(X,Y,test_size=0.2,random_state=3,stratify=Y)
```

```python
X.shape,X_train.shape,X_test.shape
```

```python
models = [LogisticRegression(max_iter=1000),SVC(kernel='linear'),KNeighborsClassifier(),RandomForestClassifier()]
```

```python
def compute_models_train_test():
  for model in models:
    model.fit(X_train,Y_train)
    testDataPrediction = model.predict(X_test)
    accuracy = accuracy_score(Y_test,testDataPrediction)
    print('model' , model ,'  =  ',accuracy)
```

```python
compute_models_train_test()
```

```python
cv_score_lr = cross_val_score(LogisticRegression(max_iter=1000),X,Y,cv=5)
```

```python
print(cv_score_lr)
mean_accuracy_r = sum(cv_score_lr)/len(cv_score_lr)
mean_accuracy_r =mean_accuracy_r*100
mean_accuracy_r =round(mean_accuracy_r,2)
print(mean_accuracy_r)
```

```python
cv_score_lr1 = cross_val_score(SVC(kernel='linear'),X,Y,cv=5)
```

```python
print(cv_score_lr1)
mean_accuracy_r1 = sum(cv_score_lr1)/len(cv_score_lr1)
mean_accuracy_r1 =mean_accuracy_r1*100
mean_accuracy_r1 =round(mean_accuracy_r1,2)
print(mean_accuracy_r1)
```

```python
cv_score_lr2 = cross_val_score(KNeighborsClassifier(),X,Y,cv=5)
mean_accuracy_r2 = sum(cv_score_lr2)/len(cv_score_lr2)
mean_accuracy_r2 =mean_accuracy_r2*100
mean_accuracy_r2 =round(mean_accuracy_r2,2)
print(mean_accuracy_r2)
```

```python
def cross_val_s():
  for model in models:
    cv_score_lr1 = cross_val_score(model,X,Y,cv=5)
    mean_accuracy_r2 = sum(cv_score_lr1)/len(cv_score_lr1)
    mean_accuracy_r2 =mean_accuracy_r2*100
    mean_accuracy_r2 =round(mean_accuracy_r2,2)
    print('model' , model ,'   == ',mean_accuracy_r2)
```

```python
cross_val_s()
```
